#### Filling missing data

1. For continous variables using the mean for filling
2. For categorical variables using the mode for filling

```python
"""
This function handles the missing data based on if it is
continous feature or categorical feature

params: 
train (dataframe)
test (datadframe)
column_list(list of features)
variable_type(str - continous or categorical)

return:
train (dataframe)
test (datadframe)
"""

def missing_data_handler(train, test, columns_list, variable_type):
    #handling missing data based on continous or categorical variable
    if variable_type == 'continous':
        for col in columns_list:
            train[col] = train[col].fillna(train[col].mean())
            test[col] = test[col].fillna(test[col].mean())
    elif variable_type == 'categorical':
        for col in columns_list:
            train[col] = train[col].fillna(train[col].mode()[0])
            test[col] = test[col].fillna(test[col].mode()[0])
    return train, test
```

```python
# list of continous variables to fill the missing data
continous_var = ['LotFrontage','GarageYrBlt','GarageCars','BsmtFinSF1','TotalBsmtSF','GarageArea',
                 'BsmtFinSF2','BsmtUnfSF','LotFrontage','GarageYrBlt','BsmtFullBath','BsmtHalfBath']
# list of categorical variables to fill the missing data
categorical_var = ['BsmtQual','BsmtCond','BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'MasVnrArea',
                   'Electrical','Exterior2nd','Exterior1st','KitchenQual','Functional','SaleType',
                   'Utilities','MSZoning','BsmtQual','BsmtCond','BsmtExposure', 'BsmtFinType1',
                   'BsmtFinType2', 'MasVnrArea', 'Electrical']

df_train, df_test = missing_data_handler(df_train, df_test, continous_var, 'continous')
df_train, df_test = missing_data_handler(df_train, df_test, categorical_var, 'categorical')
```

```python
#Check for missing values if left after the missing data treatment
df_train.isnull().sum().max(), df_test.isnull().sum().max()
```

#### Transforming the features to numeric values

Text data can't be feeded to a machine learning model, so this data preprocessing is important

```python
#Numeric Variable which are actually categorical
df_train['YrSold'] = df_train['YrSold'].apply(str)
df_train['MSSubClass'] = df_train['MSSubClass'].apply(str)

df_test['YrSold'] = df_test['YrSold'].apply(str)
df_test['MSSubClass'] = df_test['MSSubClass'].apply(str)
```

```python
#Extracting categorical features from train and test set

categorical_var_train = df_train.select_dtypes(include=[np.object])
categorical_var_test = df_train.select_dtypes(include=[np.object])
```

```python
categorical_var_train.head(2)
```

```python
categorical_var_test.head(2)
```

```python
"""
This functions encodes the categorical variables in the passed dataframe
using LabelEncodig()

params:
df (dataframe)
var_types(list of categorical variables)
return:
df (dataframe)
"""
def label_encoding(df, var_type):
    label_encoders = {}
    for col in var_type:
        label_encoders[col] = LabelEncoder()
        df[col] = label_encoders[col].fit_transform(df[col])
    return df
```

```python
#Encoding the test and train set categorical features
df_train = label_encoding(df_train, categorical_var_train)
df_test = label_encoding(df_test, categorical_var_test)
```

```python
#dependent and independent varibles data sets
#Y_df_train is dependent and X_df_train has independent data
X_df_train = df_train.drop('SalePrice', axis=1)
Y_df_train = df_train['SalePrice']
```

### Modelling the preprocessed Data 

#### Fitting the varibles into a linear regression model

```python
linear_reg_model = LinearRegression()
linear_reg_model.fit(X_df_train, Y_df_train)
```

```python
#Linear model score
linear_reg_model.score(X_df_train, Y_df_train)
```

#### Hyper tuning to improve the model
1. Gradient Boost

```python
grad_reg_model = GradientBoostingRegressor(learning_rate=0.05, n_estimators=3000, 
                                           random_state=5, max_depth=4, max_features='sqrt',
                                          min_samples_leaf=15, min_samples_split=10, loss='huber')
grad_reg_model.fit(X_df_train, Y_df_train)
```

```python
#Gradient boost model score
grad_reg_model.score(X_df_train, Y_df_train)
```

#### Hyper tuning with XGBoost
2. XGBoost

```python
xgb_reg_model = xgb.XGBRegressor(learning_rate=0.05, n_estimators=2200,
                                random_state=7, max_depth=3, min_child_weight=1.7817,
                                subsample=0.5213, silent=1, nthread=-1, reg_alpha=0.4640,
                                reg_lambda=0.8571, gamma=0.0468, colsample_bytree=0.4603)
```

```python
xgb_reg_model.fit(X_df_train, Y_df_train)
```

```python
#xbgboost score
xgb_reg_model.score(X_df_train, Y_df_train)
```

### Final Task
1. Predicting the salesPrice using the above models
2. Making a list of all the predictions
3. Creating our final output as a csv file named as 'predict_csv.csv'

```python
#list of predictions and models used to predict the Salesprice
preds = []
models = [linear_reg_model, grad_reg_model, xgb_reg_model]
for model in models:
    preds.append(model.predict(df_test))
```

```python
"""
This function does the following operations:
1. Calculated the final predictions by taking the average of 
   all the predictions by the models we used
2. Creating final output file named as 'predict_csv.csv'

params: 
sample_df(dataframe: this sample data frame show how the output should look like)
pred (list of predictions by different models)

returns:
no returns (creates the output file)
"""
def predict_csv_maker(sample_df, preds):
    final_preds = 0
    for pred in preds:
        final_preds += pred
    sample_df['SalePrice'] = (final_preds)/3
    sample_df.to_csv('Output/predict_csv.csv', index=False)
    
sample_df = pd.read_csv('Datasets/sample_submission.csv')
predict_csv_maker(sample_df, preds)
```
